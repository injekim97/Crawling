```python
import urllib.request
from bs4 import BeautifulSoup
 
url = "https://council.daegu.go.kr/kr/member/active"
req = urllib.request.Request(url)
sourcecode = urllib.request.urlopen(url).read()
soup = BeautifulSoup(sourcecode, "html.parser")
  

#  <ul class="family_site clx"> 태그 안에, <li><a href="/member/main?search_code=8013">
# -> ul 태그 안에, class 이름 으로 찾은후. find_all("li") li 태그 아래 값 추출
list = []
for href in soup.find("ul", class_="family_site clx").find_all("li"):
    #print((href.find("a")["href"][27:]))
    list.append(href.find("a")["href"][25:])

# ['', '', '8013', '8022', '1263', '8006', '8014', '8019', '8023', '8003', '8011', '8016', '1296', '8010', '8008', '1268', '8018', '8009', '8002', '8026', '8017', '8024', '8025', '8007', '1283', '8015', '8021', '8012', '8001', '8020', '8031', '8032', '', 'e=C103', 'e=C205', 'e=C215', 'e=C251', 'e=C255', 'e=C301', '', '', '', '', '', '']
#print(list)



# ★★★★★ str로 되어있는 list를 int로 바꿔서 -> 정확한 주소만 추출 하기 위해 try , except 예외처리 문 사용. ★★★★★
int_list = []
for i in list :    # 위에 찌거기 값 리스트를 반복문으로 돌려
    try :
        int_list.append(int(i))      # 새로운 배열에 int형으로 형변환해서 삽입
        
    except :  # int형으로 형변환이 안되는 것은 pass 처리 (찌거기 값 걸러짐)
        pass

print(int_list)
print(len(int_list)) #30명 
```

```python
from selenium import webdriver # 셀레니움과 웹드라이버 부르는 모듈함수 

driver = webdriver.Chrome('E:\/chromedriver.exe') # 제어하기 위해 변수 선언
driver.maximize_window() ### 창의 해상도 크기 최대화 (전체화면) ###

print("의원명","소속 정당","선거구","이메일","휴대폰 번호","사무실 번호",sep=",")

j = 1
for i in int_list:
    
    try :
        #print(str(j)+ "번째")

        driver.get(f"https://council.daegu.go.kr/member/main?search_code={i}")
        name = driver.find_element_by_xpath("//*[@id='header']/div/h1/a/span/strong") # 이름
        profile = driver.find_element_by_xpath("//*[@id='container']/div[1]/div/dl/dd/ul") # profile 전체를 가져옴
        #print(name.text[0:3],profile.text)

        
        # ★★★★★ 자 여기서, 출력된 형태에서 원하는 단어부터 출력해주기 위한 코드 ★★★★★
        flag = False    # flag 변수에 False 준 이유는, 원하는 단어('선~')을 찾게되면,
                        # True로 값을 지정해서 그 이후로 쭉 출력해주기 위한 용도
        tmp = ""
        for i in profile.text:
            # profile.text를 반복문을 통해 i로 돌림. i는 단어 한글자임 "직위에 직"
            
                
            # flag 값을 True로 줘서 그이후부터 전부 출력하기 위함
            if(flag == True):
                tmp += i

            if(flag==False and i=='선'): #'선'은 선거구에 선을 뜻함.
                tmp += i # 다음글자로 검색하기 위한 용도
                flag=True
            
             # 한글자씩 탐색하는데, \n 이라면, 뒤에 '*' 추가 -> 구분자 (CSV) 만들기 위해
            if (i == "\n"):
                tmp += '*'
                
        # 공백 2개 \n\n, 위의 조건 * 떄문에 **인데, replace로 "**" -> "*"
        tmp = tmp.replace('**','*')
        print(name.text[0:3] + tmp)
        #j += 1

    except :
        pass
```
